# Support Vector Machines Example with Scikit-Learn 

Support Vector Machines are a supervised classification model that seeks to maximise the __margin__, ie the distance between separating decision boundaries and any training data samples that are closes to this decision boundary plane. These trtaining samples that are closes to the plane are called __support vectors__.

Models with large margins tend to have lower generalisation errors, and small margins re prne to overfitting and high bias.

See Machine Learning in Python by S Raschka, Chapter 3

```python
import matplotlib.pyplot as plt
import numpy as np
import scikit_utilities as ut
from sklearn.svm import SVC
```

### Get the Iris Data

```python
X_train_std, y_train, X_test_std, y_test, X_combined_std, y_combined = ut.get_iris_data()
```

### SVM Model 

```python
svm = SVC(kernel='linear', C=1.0, random_state=0)
svm.fit(X_train_std, y_train)
```

### Plot Decision Regions


```python
plt = ut.plot_decision_regions(X_combined_std, y_combined, classifier = svm, test_idx=range(105, 150))
plt.xlabel('petal length [standardised]')
plt.ylabel('petal width [standardised]')
plt.legend(loc='upper left')
plt.show()
```

### SVM v Logistic Regression

Often yield similar results, but the logistic regression tends to be more sensitive to outliers than SVM (as it tries to maximise the conditional liklihoods of training data).however it is also a simpler model that can be implemented more easily and easily updated - good for streaming data.
