# The Iris Datset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant. One class is linearly separable from the other 2; the latter are NOT linearly separable from each other. You can get it on the following link: http://archive.ics.uci.edu/ml/datasets/Iris

It has the following attributes:

1. sepal length in cm 
2. sepal width in cm 
3. petal length in cm 
4. petal width in cm 
5. class: 
    * Iris Setosa 
    * Iris Versicolour 
    * Iris Virginica

## Getting the Dataset

```python
import numpy as np
import pandas as pd
```

### Direct from the UCI Machine Learning Repository

```python
iris_df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                      header=None, 
                      names = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Type"])
iris_df.tail()
```

Lets start off very simply by plotting the data in order as a normal line chart. We can add horizonatal lines to distinguish each type as they are ordered in the data frame. 

```python
iris_df.plot(x='Type')
plt.axvline(x=50, color='k', ls='--')
plt.axvline(x=100, color='k', ls='--')
plt.axvline(x=150, color='k', ls='--')
plt.xticks(np.arange(25, 150, 50))
```

This gives soem insight but is a little confusing. One way to see how each measurement varies by type is to use a Box and whisker Plot. This indicates range, the median and the quartiles for each measurement. I can plot this by the iris type to see how these vary for each measurement.

Here I just create a boxplot for each meaurement type, which I get as the first 4 columns in the data frame.

```python
measurements = list(iris_df.columns.values)
```

We can plot theses by the type , but this isnt the easiest way to compare. It would be much easier to compare each type for each measurement, rather than than looking at each measuremnt per type.

```python
iris_df.groupby('Type').boxplot()
```

```python
iris_df.boxplot(by='Type', layout=(2,2))
```
